```python
target_cols =[
    "last_check_date",
    "next_check_date",
    "permid",
    "clarityid",
    "issuer_name",
    "parent_company",
    "ovr_str001_active",
    "ovr_str002_active",
    "ovr_str003_active",
    "ovr_str003b_active",
    "ovr_str004_active",
    "ovr_str005_active",
    "ovr_str006_active",
    "ovr_cs001_active",
    "ovr_cs002_active",
    "ovr_cs003_active",
    "ovr_art_8_active",
]
```

```python
# let's get permids that appear more than once
dupes_permid = ovr_parents[ovr_parents.duplicated(subset="permid", keep=False)][target_cols].copy()
dupes_clarityid = ovr_parents[ovr_parents.duplicated(subset="clarityid", keep=False)][target_cols].copy()
```

```python
dupes_permid.sort_values(by="permid")
```

```python
# Sort by 'last_check_date', then 'permid', then prioritize rows with non-null 'next_check_date'
ovr_parents.sort_values(
    by=['last_check_date', 'permid', 'next_check_date'],
    ascending=[True, True, True],  # True for ascending order
    na_position='last',  # rows with NaT or NaN in next_check_date will be last
    inplace=True
)

# Reset the index after sorting
ovr_parents.reset_index(drop=True, inplace=True)
```

```python
ovr_parents_filtered = ovr_parents.drop_duplicates(subset="permid", keep="first")[target_cols].copy()
```

```python
f"the right df has {ovr_parents_filtered.shape[0]} rows while the left df has {ovr_security.shape[0]} rows"
```

```python
ovr_parents_filtered.permid = ovr_parents_filtered.permid.astype(str)
ovr_security.permid = ovr_security.permid.astype(str)
```

```python
ovr_s_updated = ovr_security.merge(ovr_parents_filtered[["last_check_date","next_check_date","permid"]], on="permid", how="left")
```

```python
ovr_s_updated.info()
```

```python
# rename column last_check_date to last_update_date and next_check_date to next_update_date in ovr_s_updated
ovr_s_updated.rename(columns={"last_check_date": "last_update_date", "next_check_date": "next_update_date"}, inplace=True)
# add empty column "creation_date" to ovr_s_updated
ovr_s_updated["creation_date"] = np.nan
# convert creation_date to datetime objet
ovr_s_updated["creation_date"] = pd.to_datetime(ovr_s_updated["creation_date"], errors="coerce")
# fill creation_date with last_update_date if creation_date is NaT
ovr_s_updated["creation_date"] = ovr_s_updated["creation_date"].fillna(ovr_s_updated["last_update_date"])
# reoderder columns in ovr_s_updated so "creation_date", "last_update_date", "next_update_date" are first
ovr_s_updated = ovr_s_updated[["creation_date", "last_update_date", "next_update_date"] + [c for c in ovr_s_updated.columns if c not in ["creation_date", "last_update_date", "next_update_date"]]]
# save ovr_s_updated to excel in "C:\Users\n740789\Downloads\overrided_db.xlsx"
ovr_s_updated.to_excel(r"C:\Users\n740789\Downloads\overrided_db.xlsx", index=False)
```

```python
OUTPU_DIR_PATH = Path(r"C:\Users\n740789\Downloads\crossreference_prepped_upload")


for cross_date, cross in cross_df_dict.items():
    print(f"\nProcessing date: {cross_date}")
    # if 'MSCI' not in columns add 'MSCI' column with np.nan values
    if 'MSCI' not in cross.columns:
        cross['MSCI'] = np.nan
    # if 'SUST' not in columns add 'SUST' column with np.nan values
    if 'SUST' not in cross.columns:
        cross['SUST'] = np.nan
    cross.rename(columns={
        'Aladdin_Issuer':'issuer',
        'Issuer_Name':'name',
        'CLARITY_AI':'Clarity'
        }, inplace=True)
    # create column 'PERMID' with copy of 'Clarity' column
    cross['PERMID'] = cross['Clarity'].copy()

    # print date and columns
    
    print(f"Columns: {cross.columns.tolist()}")
    # save to csv file with ";" as sep in OUTPU_DIR_PATH
    output_file_path = OUTPU_DIR_PATH / f"Aladdin_Clarity_Issuers_{cross_date}.csv"
    print(f"Saving to: {output_file_path}\n")
    cross.to_csv(output_file_path, sep=';', index=False, encoding='utf-8-sig')
```
